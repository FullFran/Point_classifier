# Implementación de una Red Neuronal desde Cero en Python

```python
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt

from sklearn.datasets import make_circles
```

## 1. Generación de Datos

Se utiliza la función make_circles de sklearn.datasets para generar un conjunto de datos con distribución no lineal, el cual es ideal para probar la capacidad de la red neuronal en separar patrones complejos.

- n_samples: cantidad de datos a generar.

- factor: determina la distancia entre los dos conjuntos de puntos.

- noise: introduce ruido para evitar patrones demasiado fáciles de clasificar.

```python
n = 500
p = 2

X, Y = make_circles(n_samples=n, factor=0.3, noise=0.15)
Y = Y[:, np.newaxis]
plt.scatter(X[Y[:, 0] == 0, 0], X[Y[:, 0] == 0, 1], c='skyblue')
plt.scatter(X[Y[:, 0] == 1, 0], X[Y[:, 0] == 1, 1], c='salmon')
plt.axis('equal')
plt.show()
```

## 2. Definición de la Capa Neuronal

Se implementa una clase neural_layer que representa una capa de la red neuronal.

Cada capa contiene:

- act_f: la función de activación utilizada.

- W: matriz de pesos inicializada aleatoriamente en un rango de [-1,1].

- b: vector de sesgos, también inicializado aleatoriamente.

```python
# Clase de la capa de la red
class neural_layer():
    def __init__(self, n_conn, n_neur, act_f):
        self.act_f = act_f
        self.b = np.random.rand(1, n_neur) * 2 - 1
        self.W = np.random.rand(n_conn, n_neur) * 2 - 1
```

## 3. Funciones de Activación

Se implementa la función sigmoide y ReLU, así como sus derivadas.

```python
# Funciones de activación:
sigm = (lambda x: 1 / (1 + np.exp(-x)),
        lambda x: x * (1 - x)
        )


def relu(x): return np.maximum(0, x)


_x = np.linspace(-5, 5, 100)

plt.plot(_x, sigm[0](_x))
plt.show()
```

## 4. Creación de la Red Neuronal

Se define la estructura de la red mediante create_nn, que recibe como entrada una lista topology que indica el número de neuronas por capa.

```python
# Creamos nuestra red neuronal
def create_nn(topology, act_f):
    nn = []
    for l, layer in enumerate(topology[:-1]):
        nn.append(neural_layer(topology[l], topology[l+1], act_f))

    return nn
```

# 5. Entrenamiento de la Red Neuronal

- Definimos la red con su topología
- Definimos la función de coste en este caso la Binary Cross Entropy para el problema de clasificación.

### Forward Pass

Se calculan las activaciones de cada capa aplicando la transformación lineal: $Z = XW + bA = f(Z)$

### Backpropagation (Backward Pass)

Se calculan los gradientes y se actualizan los pesos mediante el descenso de gradiente

### Actualización de Pesos y Sesgos

Se actualizan los pesos utilizando los gradientes obtenidos

```python
topology = [p, 4, 8, 1]
neural_net = create_nn(topology, sigm)
l2_cost = (lambda Yp, Yr: np.mean(Yp - Yr) ** 2,
           lambda Yp, Yr: (Yp - Yr))

# vamos a usar como función de coste la bynary cross entropy
l2_cost = (lambda Yp, Yr: np.mean((Yr * np.log(Yp) + (1 - Yr) * np.log(1 - Yp))),
           lambda Yp, Yr: (Yp - Yr))


def train(neural_net, X, Y, l2_cost, lr=0.5, train=True):
    out = [(None, X)]
    # fordward pass:
    for l, layer in enumerate(neural_net):
        z = out[-1][1] @ neural_net[l].W + neural_net[l].b  # la 'cajita'
        a = neural_net[l].act_f[0](z)                       # aplicamos la función a la cajita.
        out.append((z, a))

    # backpropagation:
    if train:
        deltas = []
        for l in reversed(range(0, len(neural_net))):

            z = out[l+1][0]
            a = out[l+1][1]

            if l == len(neural_net) - 1:
                # delta ultima capa
                deltas.insert(0, l2_cost[1](a, Y) * neural_net[l].act_f[1](a))

            else:
                deltas.insert(0, deltas[0] @ _W.T * neural_net[l].act_f[1](a))
                # delta capa

            _W = neural_net[l].W
            # Gradient descent
            neural_net[l].b = neural_net[l].b - \
                np.mean(deltas[0], axis=0, keepdims=True) * lr
            neural_net[l].W = neural_net[l].W - out[l][1].T @ deltas[0] * lr
    return out[-1][1]


train(neural_net, X, Y, l2_cost, 0.5)
print('')
```

## Mostramos los resultados

```python
import time
from IPython.display import clear_output

neural_n = create_nn(topology, sigm)  # creamos la red

loss = []

for i in range(1500):

    pY = train(neural_n, X, Y, l2_cost, lr=0.05)  # damos un paso del entrenamiento


    # Cada 25 pasos mostramos el resultado.
    if i % 25 == 0:

        loss.append(abs(l2_cost[0](pY, Y)))

        res = 50

        _x0 = np.linspace(-1.5, 1.5, res)
        _x1 = np.linspace(-1.5, 1.5, res)

        _Y = np. zeros((res, res))

        for i0, x0 in enumerate(_x0):
            for i1, x1 in enumerate(_x1):
                _Y[i0, i1] = train(neural_n, np.array(
                    [[x0, x1]]), Y, l2_cost, train=False)[0][0]

        plt.pcolormesh(_x0, _x1, _Y, cmap='coolwarm')
        plt.axis('equal')

        plt.scatter(X[Y[:, 0] == 0, 0], X[Y[:, 0] == 0, 1], c='skyblue')
        plt.scatter(X[Y[:, 0] == 1, 0], X[Y[:, 0] == 1, 1], c='salmon')

        clear_output(wait=True)
        plt.show()
        plt.loglog(range(len(loss)), loss)
        plt.show()
        time.sleep(0.5)
```
